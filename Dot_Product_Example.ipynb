**Dot Product Example - Ice-cream sales**

```python
import numpy as np
import pandas as pd
```

```python
np.random.seed(0)
# Number of ice-creams sold
sales_amt = np.random.randint(15, size=(5,3))
# Size is mentioned as 5 rows which specifies '5 days in a week' and 3 columns specifies 'Flavours of Ice-creams'
sales_amt
```

```python
# Creating weekly_sales DataFrame
weekly_sales = pd.DataFrame(sales_amt,
                            index=["Mon", "Tues", "Wed", "Thurs", "Fri"],
                            columns = ["Black currant", "Chocolate", "Badam Kulfi"])
weekly_sales
```

```python
# Creating Prices array
prices = np.array([25, 20, 30])
prices
```

```python
# Creating ice-cream_prices DataFrame
icecream_prices = pd.DataFrame(prices.reshape(1, 3),
                             index = ["Price"],
                             columns = ["Black currant", "Chocolate", "Badam Kulfi"])
icecream_prices
```

```python
prices.shape, sales_amt.shape #Both didn't align to perform the dot operation
```

```python
total_sales = prices.dot(sales_amt.T)
total_sales
```

```python
#Create daily_sales
icecream_prices
```

```python
weekly_sales
```

```python
icecream_prices.shape, weekly_sales.shape
```

```python
daily_sales = icecream_prices.dot(weekly_sales.T)
daily_sales
```

```python
weekly_sales
```

```python
weekly_sales["Total (Rs.)"] = daily_sales.T
# Shape of daily_sales is (1, 5)
# We must transpose to add in the weekly_sales report
weekly_sales
```
